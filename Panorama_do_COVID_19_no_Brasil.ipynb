# **Panorama do COVID-19 no Brasil**

COVID-19 é uma doença infecciosa causada pelo novo coronavírus (SARS-CoV-2).

Em 31 de dezembro de 2019, a **Organização Mundial da Saúde (OMS)** recebeu o alerta sobre uma grande quantidade de casos de pneumonia na cidade de Wuhan, província de Hubei, na **República Popular da China**. O causador dos casos de pneumonia foi identificado como uma nova cepa, **um novo tipo de coronavírus** que anteriormente ainda não havia sido registrada a ocorrência em seres humanos.

Com a disseminação do novo coronavírus em larga escala pelo mundo, transmitido de pessoa para pessoa, a OMS declarou em 30 de janeiro de 2020, que o surto do novo coronavírus constitui uma **Emergência de Saúde Pública de Importância Internacional (ESPII), significando como o mais alto nível de alerta da Organização, de acordo com o Regulamento Sanitário Internacional.** Com o intuito de interromper a propagação do vírus, a decisão da OMS buscou aprimorar a coordenação, a cooperação e a solidariedade global em combate a nova doença.

## Obtenção dos Dados


Os dados utilizados para esta análise foram obtidos através do portal [Our World in Data](https://ourworldindata.org/coronavirus) que consiste em uma fonte aberta e contêm uma coleção de dados completa sobre o COVID-19 por todo o mundo, a qual é atualizada diariamente. Segundo o portal os dados são mantidos por uma equipe do [Centro de Ciência e Engenharia de Sistemas (CSSE)](https://systems.jhu.edu/) da [Universidade Johns Hopkins (JHU)](https://coronavirus.jhu.edu/) a qual coletam informações públicas de fontes oficiais.

A fonte dos dados é algo extremamente importante para realização de uma análise de uma nova doença como é o caso do Coronavírus, pois o principal desafio é a confiabilidade e integridade dos dados, visto que existem várias fontes, como; dados do governo federal, governo estaduais, secretarias de saúde, dados da imprensa, entre outros, e isto pode acarretar em conflitos de informações.  

*Esta análise contêm os dados publicados a partir da data de 24-02-2020 até 14-06-2021.*

### **Dicionário de Variáveis**

Variáveis contidas no *dataset* e suas especificações:

* `iso_code` - ISO 3166-1 alfa-3 - códigos de três letras do país
* `continent` - Continente da localização geográfica
* `location` - Localização geográfica
* `date` - Data de observação
* `total_cases` - Total de casos confirmados de COVID-19
* `new_cases` - Novos casos confirmados de COVID-19
* `new_cases_smoothed` - 	Novos casos confirmados de COVID-19 (corrigido em 7 dias)
* `total_deaths` - Total de mortes atribuídas a COVID-19
* `new_deaths` - Novas mortes atribuídas a COVID-19
* `new_deaths_smoothed` - Novas mortes atribuídas a COVID-19 (suavizado por 7 dias)
* `total_cases_per_million` - Total de casos confirmados de COVID-19 por 1.000.000 de pessoas
* `new_cases_per_million` - Novos casos confirmados de COVID-19 por 1.000.000 de pessoas
* `new_cases_smoothed_per_million` - Novos casos confirmados de COVID-19 (corrigido em 7 dias) por 1.000.000 de pessoas
* `total_deaths_per_million` - 	Total de mortes atribuídas a COVID-19 por 1.000.000 de pessoas
* `new_deaths_per_million` - Novas mortes atribuídas a COVID-19 por 1.000.000 de pessoas
* `new_deaths_smoothed_per_million` - 	Novas mortes atribuídas a COVID-19 (suavizado por 7 dias) por 1.000.000 de pessoas
* `reproduction_rate` - Estimativa em tempo real da taxa efetiva de reprodução (R) do COVID-19. Veja https://github.com/crondonm/TrackingR/tree/main/Estimates-Database
* `icu_patients` - 	Número de pacientes COVID-19 em unidades de terapia intensiva (UTI) em um determinado dia
* `icu_patients_per_million` - 	Número de pacientes COVID-19 em unidades de terapia intensiva (UTI) em um determinado dia por 1.000.000 de pessoas
* `hosp_patients` - Número de pacientes COVID-19 no hospital em um determinado dia
* `hosp_patients_per_million` - Número de pacientes COVID-19 no hospital em um determinado dia por 1.000.000 de pessoas
* `weekly_icu_admissions` - Número de pacientes COVID-19 recém-admitidos em unidades de terapia intensiva (UTI) em uma determinada semana
* `weekly_icu_admissions_per_million` - Número de pacientes COVID-19 recém-admitidos em unidades de terapia intensiva (UTI) em uma determinada semana por 1.000.000 de pessoas
* `weekly_hosp_admissions` - Número de pacientes COVID-19 recém-admitidos em hospitais em uma determinada semana
* `weekly_hosp_admissions_per_million` - Número de pacientes COVID-19 recém-admitidos em hospitais em uma determinada semana por 1.000.000 de pessoas
* `total_tests` - Total de testes para COVID-19
* `new_tests` - Novos testes para COVID-19 (calculado apenas para dias consecutivos)
* `total_tests_per_thousand` - Total de testes para COVID-19 por 1.000 pessoas
* `new_tests_per_thousand` - Novos testes para COVID-19 por 1.000 pessoas
* `new_tests_smoothed` - 	Novos testes para COVID-19 (suavizado por 7 dias). Para países que não relatam dados de teste diariamente, presumimos que os testes mudaram igualmente diariamente em todos os períodos em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que é então calculada em média ao longo de uma janela contínua de 7 dias
* `new_tests_smoothed_per_thousand` -	Novos testes para COVID-19 (suavizado por 7 dias) por 1.000 pessoas
* `positive_rate` - A parcela de testes COVID-19 que são positivos, dada como uma média contínua de 7 dias (este é o inverso de tests_per_case)
* `tests_per_case` - Testes conduzidos por novo caso confirmado de COVID-19, dado como uma média contínua de 7 dias (este é o inverso de Positive_rate)
* `tests_units` - Unidades usadas pelo local para relatar seus dados de teste
* `total_vaccinations` - Número total de doses de vacinação COVID-19 administradas
* `people_vaccinated` - Número total de pessoas que receberam pelo menos uma dose de vacina
* `people_fully_vaccinated` - Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação
* `new_vaccinations` - Novas doses de vacinação COVID-19 administradas (calculadas apenas para dias consecutivos)
* `new_vaccinations_smoothed` - Novas doses de vacinação COVID-19 administradas (suavização de 7 dias). Para os países que não relatam dados de vacinação diariamente, presumimos que a vacinação mudou igualmente diariamente em qualquer período em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que é então calculada em média ao longo de uma janela contínua de 7 dias
* `total_vaccinations_per_hundred` - Número total de doses de vacinação COVID-19 administradas por 100 pessoas na população total
* `people_vaccinated_per_hundred` -	Número total de pessoas que receberam pelo menos uma dose de vacina por 100 pessoas na população total
* `people_fully_vaccinated_per_hundred` - Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação por 100 pessoas na população total
* `new_vaccinations_smoothed_per_million` - Novas doses de vacinação COVID-19 administradas (suavizadas em 7 dias) por 1.000.000 de pessoas na população total
* `stringency_index` - Índice de Restrição de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamentos de escolas, fechamentos de locais de trabalho e proibições de viagens, redimensionado para um valor de 0 a 100 (100 = resposta mais estrita)
* `population` - População em 2020
* `population_density` - 	Número de pessoas dividido por área terrestre, medido em quilômetros quadrados, ano mais recente disponível
* `median_age` - Idade média da população, projeção da ONU para 2020
* `aged_65_older` - Parcela da população com 65 anos ou mais, ano mais recente disponível
* `aged_70_older` -	Parcela da população com 70 anos ou mais em 2015
* `gdp_per_capita` - Produto interno bruto em paridade de poder de compra (dólares internacionais constantes de 2011), ano mais recente disponível
* `extreme_poverty` - Parcela da população que vive em pobreza extrema, ano mais recente disponível desde 2010
* `cardiovasc_death_rate` - Taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de mortes por 100.000 pessoas)
* `diabetes_prevalence` - Prevalência de diabetes (% da população de 20 a 79 anos) em 2017
* `female_smokers` - Parcela de mulheres que fumam, ano mais recente disponível
* `male_smokers` - Proporção de homens que fumam, ano mais recente disponível
* `handwashing_facilities` - Parcela da população com instalações básicas de lavagem das mãos no local, ano mais recente disponível
* `hospital_beds_per_thousand` - Leitos hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010
* `life_expectancy` - Expectativa de vida ao nascer em 2019
* `human_development_index` - Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano - uma vida longa e saudável, conhecimento e um padrão de vida decente. Valores para 2019, importados de http://hdr.undp.org/en/indicators/137506
* `excess_mortality` - Excesso de mortalidade P-score para todas as idades; consulte https://github.com/owid/covid-19-data/tree/master/public/data/excess_mortality para a definição do P-score e como o calculamos

### **Análise Exploratória para o Mundo**

* ***Visualizar um gráfico para ver a evolução do total de Casos e o total de Mortes por COVID-19 no mundo***

Os gráficos a seguir apresentam a evolução do total de casos de infecção pelo COVID-19 e o total de mortes que foram registradas pelo mundo durante o período de 24-02-2020 até 14-06-2021. 
* O gráfico do **total de casos** possui os valores representados em **centenas de milhão** (mais de 175 milhões). 
* O gráfico do **total de mortes** possui os valores representados em **milhões** (mais de 35 milhões).

```
# criar uma cópia do dataset
df_copy = pd.read_csv(data)
# converte a coluna date em datetime
df_copy.date = pd.to_datetime(df_copy.date)
# adicionar a coluna date como index
df_copy.set_index('date', inplace=True)
```

```
# exibir a evoluçao do total de casos confirmados e de mortes pelo mundo
# declarar os objetos do gráfico
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

# plotar total de casos
df_copy.loc['2020-02-24':'2021-06-14', "total_cases"].plot(ax=ax1)
ax1.set_title("Gráfico do Total de Casos")
ax1.set_ylabel("Casos Confirmados por Centenas de Milhão")
ax1.set_xlabel("Data")

# plotar total de mortes
df_copy.loc['2020-02-24':'2021-06-14', "total_deaths"].plot(ax=ax2)
ax2.set_title("Gráfico do Total de Mortes")
ax2.set_ylabel("Mortes Registradas em Milhões")
ax2.set_xlabel("Data")

plt.tight_layout()
```

* ***Visualizar em gráfico para ver a evolução do total de Casos e o total de Mortes distribuídos por continentes***

Os gráficos a diante apresentam a distribuição do total dos casos confirmados e de mortes por continentes. 
* O gráfico do **total de casos** possui os valores representados em **milhões**. 
* O gráfico do **total de mortes** possui os valores representados em **centenas de milhar**.

Percebe-se um contraste em relação aos casos confirmados e as mortes entre os continentes. Enquanto o continente Asiático possui o maior numero de casos confirmados, ultrapassando o total de 5 milhões, os casos de mortes apresentam uma taxa bem menor em comparação aos continentes Europeu, Norte e Sul Americanos, estes que respectivamente possuem uma taxa de casos confirmados de aproximadamente 5, 4 e 3 milhões. A Europa é o continente com maior numero de mortes.

```
# criar uma cópia do dataset pelos continentes
df_continents = pd.DataFrame();
df_continents = df_copy.loc[(df_copy['location'] == 'Asia') | (df_copy['location'] == 'Africa') | (df_copy['location'] == 'Oceania') | (df_copy['location'] == 'South America') | (df_copy['location'] == 'North America') | (df_copy['location'] == 'Europe')].copy()
```

```
# exibir gráficos para o total de casos confirmados e mortes por continentes
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,6))
ax1.bar('location', 'total_cases', data=df_continents)
ax1.set_title("Gráfico do Total de Casos por Continentes")
ax1.set_ylabel("Casos Confirmados em Milhões")

ax2.bar('location', 'total_deaths', data=df_continents)
ax2.set_title("Gráfico do Total de Mortes por Continentes")
ax2.set_ylabel("Mortes Registradas em Centenas de Milhar")

fig.show();
```

* ***Visualizar em gráfico o total de Casos em comparação a população por continentes***

A seguir o gráfico apresenta a distribuição do total de casos confirmados em relação ao total da população de cada continente.

```
# exibir gráfico da população e total de casos por continentes
fig, ax = plt.subplots(figsize=(10,6))
ax.bar('location', 'population', data=df_continents, color = 'blue', label="População Total")
ax.bar('location', 'total_cases', data=df_continents, color = 'red', label="Casos Confirmados")
ax.set_title("Gráfico da População e o Total de Casos por Continentes")
ax.set_ylabel("Casos Confirmados/População em Milhões")

fig.legend()
fig.show()
```

* ***Ver para a data mais atual, quais os países que têm mais casos e mortes.***

A seguir, está listado em ordem decrescente os 5 países que possuem o maior numero de mortes derivadas do COVID-19. Os Estados Unidos é o país que tem o maior índice de mortes, e em segundo lugar temos o Brasil, com aproximadamente 500 mil mortes.

```
# exibir os países com mais mortes
df_copy = df.loc[df.continent.isnull() == False].copy()
df_copy.loc[df_copy.date == '2021-06-14'].sort_values(by="total_deaths", ascending=False)[0:5]
```

* ***Visualizar em gráfico de barras os 5 países com mais casos confirmados e mais mortes***

Nos gráficos a diante percebemos o índice de casos confirmados e o índice de mortes dos países que estão entre os cinco primeiros em relação a cada categoria, de casos confirmados e mortes.

* Os Estados Unidos é o país que possui os maiores índices, tanto para casos confirmados como para mortes. A Índia é o segundo em relação ao índice de contaminação, mas fica atrás do Brasil em relação aos casos de mortes, sendo que o Brasil é o terceiro colocado em relação aos casos confirmados.

```
# plotar gráfico de barras para os 5 países
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

top5_paises_casos = df_copy.loc[df_copy.date == '2021-06-14', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False)[0:5]
ax1.bar('location', 'total_cases', data=top5_paises_casos)
ax1.set_title("Gráfico do Total de Casos por Países")
ax1.set_ylabel("Casos Confirmados em Milhões")

top5_paises_mortes = df_copy.loc[df_copy.date == '2021-06-14', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]
ax2.bar('location', 'total_deaths', data=top5_paises_mortes)
ax2.set_title("Gráfico do Total de Mortes por Países")
ax2.set_ylabel("Mortes Registradas")
fig.show();
```

* **Identificar data do primeiro caso de morte no Brasil**

No Brasil, a quantidade de casos confirmados e casos de mortes aumentaram significamente no decorrer dos tempos, é notável nos gráficos a seguir em escala linear o crescimento contínuo tanto para casos confirmados quanto para mortes, estas que se aproximam do total de 500 mil.

```
# declarar os objetos do gráfico
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

# plotar total de casos em escala linear
df_brasil_copy.loc['2020-02-24':'2021-06-14', "total_cases"].plot(ax=ax1)
ax1.set_title("Gráfico do Total de Casos")
ax1.set_ylabel("Casos Confirmados em Milhão")
ax1.set_xlabel("Data")

# plotar total de mortes em escala linear
df_brasil_copy.loc['2020-02-24':'2021-06-14', "total_deaths"].plot(ax=ax2)
ax2.set_title("Gráfico do Total de Mortes")
ax2.set_ylabel("Mortes Registradas")
ax2.set_xlabel("Data")

plt.tight_layout()
```

* **Visualizar gráficos de linha para evolução dos casos CONFIRMADOS e casos de MORTES por COVID-19 no Brasil em ESCALA LOGARÍTMICA**

Os gráficos a seguir apresentam a evolução dos casos de covid-19 no Brasil em escala logarítmica.

```
# declarar os objetos do gráfico
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

# plotar total de casos em escala logarítmica
df_brasil_copy.loc['2020-02-24':'2021-06-14', "total_cases"].plot(ax=ax1)
ax1.set_title("Gráfico do Total de Casos (Escala logarítmica)")
ax1.set_ylabel("Casos Confirmados em Milhão")
ax1.set_yscale('log')
ax1.set_xlabel("Data")

# plotar total de mortes em escala logarítmica
df_brasil_copy.loc['2020-02-24':'2021-06-14', "total_deaths"].plot(ax=ax2)
ax2.set_title("Gráfico do Total de Mortes (Escala logarítmica)")
ax2.set_ylabel("Mortes Registradas")
ax2.set_yscale('log')
ax2.set_xlabel("Data")

plt.tight_layout()
```
